```python
import pandas as pd

avg_df = pd.read_csv(r'datasets\avg_stats.csv').set_index('#')

tot_df = pd.read_csv(r'datasets\total_stats.csv').set_index('#')

per36_df = pd.read_csv(r'datasets\per36_stats.csv').set_index('#')

adv_df = pd.read_csv(r'datasets\adv_stats_cleaned.csv').set_index('#')

misc_df = pd.read_csv(r'datasets\misc_stats_cleaned.csv').set_index('#')

```

```python
avg_tot_df = avg_df.merge(tot_df[tot_df.columns.difference(avg_df.columns).tolist() + ['Player']], how = 'inner', on = ['Player'])
avg_tot_df
```

```python
avg_tot_per36_df = avg_tot_df.merge(per36_df[per36_df.columns.difference(avg_tot_df.columns).tolist() + ['Player']], how = 'inner', on = ['Player'])
avg_tot_per36_df
```

```python
avg_tot_per36_adv_df = avg_tot_per36_df.merge(adv_df[adv_df.columns.difference(avg_tot_per36_df.columns).tolist() + ['Player']], how = 'inner', on = ['Player'])
avg_tot_per36_adv_df
```

```python
avg_tot_per36_adv_misc_df = avg_tot_per36_adv_df.merge(misc_df[misc_df.columns.difference(avg_tot_per36_adv_df.columns).tolist() + ['Player']], how = 'inner', on = ['Player'])
avg_tot_per36_adv_misc_df
```

```python
avg_tot_per36_adv_misc_df.to_csv(path_or_buf=r'datasets\compiled_stats.csv', index=False)
```
